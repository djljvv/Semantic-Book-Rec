```python
import numpy as np
#extract max emotional prob for each description
emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"] #key in dictionary then into columns
isbn = [] #key to merge everything back into data frame
emotion_scores = {label:[] for label in emotion_labels} #dictionary comprehension to contain all score for all description for each label

#extracts score after looping through sorted predictions using right label
def calculate_max_emotion_scores(predictions):
    per_emotion_scores = {label: [] for label in emotion_labels}
    for prediction in predictions:
        # 🔑 build a dict {label: score} instead of sorting alphabetically
        pred_dict = {p["label"]: p["score"] for p in prediction}
        for label in emotion_labels:
            per_emotion_scores[label].append(float(pred_dict.get(label, 0.0)))
    return {label: float(np.max(scores)) for label, scores in per_emotion_scores.items()}

```

```python
for i in range(10):
    isbn.append(books["isbn13"][i])
    sentences = books["description_x"][i].split(".")
    predictions = classifier(sentences) #looping through setences in descriptions to pass into classifier
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels: #takes the max result and append it to emotion scores dictionary
        emotion_scores[label].append(max_scores[label])
```

```python
from tqdm import tqdm

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"] #key in dictionary then into columns
isbn = [] #key to merge everything back into data frame
emotion_scores = {label:[] for label in emotion_labels}

for i in tqdm(range(len(books))):
    isbn.append(books["isbn13"][i])
    sentences = books["description_x"][i].split(".")
    predictions = classifier(sentences) #looping through setences in descriptions to pass into classifier
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels: #takes the max result and append it to emotion scores dictionary
        emotion_scores[label].append(max_scores[label])
```

```python
emotions_df = pd.DataFrame(emotion_scores)
emotions_df["isbn13"] = isbn
```

```python
emotions_df.head()
```

```python
books = pd.merge(books, emotions_df, on="isbn13")
```
