```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Colums")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np
books["missing_description"] = np.where(books["description"].isna(), 1, 0) #creates a column in books database checking if there is a description using isna, and setting output of isna to 1,0 instead of T/F
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"] #list of subset of columns we want from books data set

correlation_matrix = books[columns_of_interest].corr(method="spearman")#creating a correlation matrix using pd frame columns, pearson is better for continuous variables where spearman is better for binary or non continuous variables

sns.set_theme(style="white")
plt.figure(figsize = (8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f",cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show() #shows how everythign works together, anything that is close to 0 is weak correlation, 1 or -1 is strong
```

```python
book_missing = books[~(books["description"].isna()) &
~(books["published_year"].isna()) &
~(books["average_rating"].isna()) &
      ~(books["num_pages"].isna())
]#excludes all entries missing any of these
"""books[(books["description"].isna()) |
(books["published_year"].isna()) |
(books["average_rating"].isna()) |
      (books["num_pages"].isna())
]

This finds all the columns that have a missing description OR any of the other things missing"""
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False) #reset index makes it so categories are each put into a column, and then counts the amount per category by descending, allows us to see what in the data we need to remove
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
#split the descriptions to give us the number of words
```

```python
book_missing
```

```python
book_missing.loc[book_missing["words_in_description"].between(15, 24), "description"]# you have to filter the data in all categories in order to ensure that it won't produce innaccurate reccomendations
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"], book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join, axis=1) #attaches the isbn at the start
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1).to_csv("books_cleaned.csv", index=False)#drops the unnecessary columns so we can properly use the data
)
```
